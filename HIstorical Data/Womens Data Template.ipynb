```python
import pandas as pd

master = pd.read_csv('mastering_wncaat_teams_historical.csv')
summary = pd.read_csv('raw_herhoops_summary_historical.csv')
summaryadv = pd.read_csv('raw_herhoops_summaryadv_historical.csv')
advanced = pd.read_csv('raw_herhoops_advanced_historical.csv')
opponent = pd.read_csv('raw_herhoops_opponent_historical.csv')
```

## Teams

```python
summary = summary[['her_hoops_id', 'PTS', 'OPP PTS', 'Net PTS', 'Off Rtg', 'Def Rtg', 'Net Rtg', 'pace']]
summary = summary.rename(columns={
    'PTS': 'ppg', 
    'OPP PTS': 'opp_ppg', 
    'Net PTS': 'net_ppg', 
    'Off Rtg': 'off_rtg', 
    'Def Rtg': 'def_rtg', 
    'Net Rtg': 'net_rtg'
})
summaryadv = summaryadv[['her_hoops_id', 'HHS Rtg', 'HHS Off Rtg', 'HHS Def Rtg', 'Simple RPI']]
summaryadv = summaryadv.rename(columns={
    'HHS Rtg': 'hhs_net_rtg',
    'HHS Off Rtg': 'hhs_off_rtg',
    'HHS Def Rtg': 'hhs_def_rtg',
    'Simple RPI': 'rpi'
})
advanced = advanced[['her_hoops_id', 'PPP', 'ppsa', 'EFG%', 'FT Rate', '3P Rate',
                     '% Pts FT', '% Pts 2P', '% Pts 3P', 'ORB%', 'DRB%', 'TRB%', 
                     'AST%', 'TOV%', 'AST/TOV', 'STL%', 'BLK%', 'PF%']]
advanced = advanced.rename(columns={
    'PPP': 'ppp', 
    'EFG%': 'efg%', 
    'FT Rate': 'ftr', 
    '3P Rate': '3pr',
    '% Pts FT': '%pts_ft', 
    '% Pts 2P': '%pts_2p', 
    '% Pts 3P': '%pts_3p', 
    'ORB%': 'orb%', 
    'DRB%': 'drb%', 
    'TRB%': 'trb%', 
    'AST%': 'ast%', 
    'TOV%': 'tov%', 
    'AST/TOV': 'ast_tov', 
    'STL%': 'stl%', 
    'BLK%': 'blk%', 
    'PF%': 'pf%'
})
opponent = opponent[['her_hoops_id', 'PPP', 'PPSA', 'EFG%', 'FT Rate', '3P Rate',
                     '% PTS FT', '% PTS 2P', '% PTS 3P', 'ORB%', 'DRB%', 'TRB%', 
                     'AST%', 'TOV%', 'AST/TOV', 'STL%', 'BLK%', 'PF%']]
opponent = opponent.rename(columns={
    'PPP': 'pppd',
    'PPSA': 'opp_ppsa',
    'EFG%': 'efgd%', 
    'FT Rate': 'ftrd', 
    '3P Rate': '3prd',
    '% PTS FT': '%pts_ftd', 
    '% PTS 2P': '%pts_2pd', 
    '% PTS 3P': '%pts_3pd', 
    'ORB%': 'orbd%', 
    'DRB%': 'drbd%', 
    'TRB%': 'trbd%', 
    'AST%': 'astd%', 
    'TOV%': 'tovd%', 
    'AST/TOV': 'opp_ast_tov', 
    'STL%': 'opp_stl%', 
    'BLK%': 'opp_blk%', 
    'PF%': 'opp_pf%'
})                      
```

```python
print(master.shape)
join_1 = pd.merge(master, summary, on='her_hoops_id', how='inner')
print(join_1.shape)
join_2 = pd.merge(join_1, summaryadv, on='her_hoops_id', how='inner')
print(join_2.shape)
join_3 = pd.merge(join_2, advanced, on='her_hoops_id', how='inner')
print(join_3.shape)
join_4 = pd.merge(join_3, opponent, on='her_hoops_id', how='inner')
print(join_4.shape)
```

```python
join_4.columns.to_list()
```

```python
join_4['opp_ppg'] = 120 - join_4['opp_ppg']
join_4['def_rtg'] = 120 - join_4['def_rtg']
join_4['hhs_def_rtg'] = 120 - join_4['hhs_def_rtg']
join_4['pppd'] = 1.2 - join_4['pppd']
join_4['opp_ppsa'] = 1.2 - join_4['opp_ppsa']
join_4['efgd%'] = 1.00 - join_4['efgd%']
```

```python
join_6 = pd.merge(games, join_4, left_on='low_team_id', right_on='team_id', how='inner')
print(join_6.shape)
```

```python
join_6 = join_6.rename(columns={
 'ppg': 'low_ppg',
 'opp_ppg': 'low_opp_ppg',
 'net_ppg': 'low_net_ppg',
 'off_rtg': 'low_off_rtg',
 'def_rtg': 'low_def_rtg',
 'net_rtg': 'low_net_rtg',
 'pace': 'low_pace',
 'hhs_net_rtg': 'low_hhs_net_rtg',
 'hhs_off_rtg': 'low_hhs_off_rtg',
 'hhs_def_rtg': 'low_hhs_def_rtg',
 'rpi': 'low_rpi',
 'ppp': 'low_ppp',
 'ppsa': 'low_ppsa',
 'efg%': 'low_efg%',
 'ftr': 'low_ftr',
 '3pr': 'low_3pr',
 '%pts_ft': 'low_%pts_ft',
 '%pts_2p': 'low_%pts_2p',
 '%pts_3p': 'low_%pts_3p',
 'orb%': 'low_orb%',
 'drb%': 'low_drb%',
 'trb%': 'low_trb%',
 'ast%': 'low_ast%',
 'tov%': 'low_tov%',
 'ast_tov': 'low_ast_tov',
 'stl%': 'low_stl%',
 'blk%': 'low_blk%',
 'pf%': 'low_pf%',
 'pppd': 'low_pppd',
 'opp_ppsa': 'low_opp_ppsa',
 'efgd%': 'low_efgd%',
 'ftrd': 'low_ftrd',
 '3prd': 'low_3prd',
 '%pts_ftd': 'low_%pts_ftd',
 '%pts_2pd': 'low_%pts_2pd',
 '%pts_3pd': 'low_%pts_3pd',
 'orbd%': 'low_orbd%',
 'drbd%': 'low_drbd%',
 'trbd%': 'low_trbd%',
 'astd%': 'low_astd%',
 'tovd%': 'low_tovd%',
 'opp_ast_tov': 'low_opp_ast_tov',
 'opp_stl%': 'low_opp_stl%',
 'opp_blk%': 'low_opp_blk%',
 'opp_pf%': 'low_opp_pf%'
})
```

```python
games_2 = games[['game_id', 'year', 'round', 'region', 'win']]
print(games_2.shape)
high = join_5[['game_id', 'high_team_id', 'high_ppg', 'high_opp_ppg',
       'high_net_ppg', 'high_off_rtg', 'high_def_rtg', 'high_net_rtg',
       'high_pace', 'high_hhs_net_rtg', 'high_hhs_off_rtg', 'high_hhs_def_rtg',
       'high_rpi', 'high_ppp', 'high_ppsa', 'high_efg%', 'high_ftr',
       'high_3pr', 'high_%pts_ft', 'high_%pts_2p', 'high_%pts_3p', 'high_orb%',
       'high_drb%', 'high_trb%', 'high_ast%', 'high_tov%', 'high_ast_tov',
       'high_stl%', 'high_blk%', 'high_pf%', 'high_pppd', 'high_opp_ppsa',
       'high_efgd%', 'high_ftrd', 'high_3prd', 'high_%pts_ftd',
       'high_%pts_2pd', 'high_%pts_3pd', 'high_orbd%', 'high_drbd%',
       'high_astd%', 'high_tovd%', 'high_opp_ast_tov',
       'high_opp_stl%', 'high_opp_blk%', 'high_opp_pf%']]
print(high.shape)
low = join_6[['game_id', 'low_team_id', 'low_ppg', 'low_opp_ppg',
       'low_net_ppg', 'low_off_rtg', 'low_def_rtg', 'low_net_rtg', 'low_pace',
       'low_hhs_net_rtg', 'low_hhs_off_rtg', 'low_hhs_def_rtg', 'low_rpi',
       'low_ppp', 'low_ppsa', 'low_efg%', 'low_ftr', 'low_3pr', 'low_%pts_ft',
       'low_%pts_2p', 'low_%pts_3p', 'low_orb%', 'low_drb%', 'low_trb%',
       'low_ast%', 'low_tov%', 'low_ast_tov', 'low_stl%', 'low_blk%', 'low_pf%',
       'low_pppd', 'low_opp_ppsa', 'low_efgd%', 'low_ftrd', 'low_3prd',
       'low_%pts_ftd', 'low_%pts_2pd', 'low_%pts_3pd', 'low_orbd%', 'low_drbd%',
       'low_astd%', 'low_tovd%', 'low_opp_ast_tov', 'low_opp_stl%',
       'low_opp_blk%', 'low_opp_pf%']]
print(low.shape)
games_3 = pd.merge(games_2, high, on='game_id', how='inner')
print(games_3.shape)
games_4 = pd.merge(games_3, low, on='game_id', how='inner')
print(games_4.shape)
```

```python
matchups                =   games_4[['game_id', 'year', 'round', 'region', 'high_team_id', 'low_team_id', 'win']]
matchups['ppg']         =   games_4['high_ppg']         - games_4['low_opp_ppg']
matchups['opp_ppg']     =   games_4['high_opp_ppg']     - games_4['low_ppg']
matchups['net_ppg']     =   games_4['high_net_ppg']     - games_4['low_net_ppg']
matchups['off_rtg']     =   games_4['high_off_rtg']     - games_4['low_def_rtg']
matchups['def_rtg']     =   games_4['high_def_rtg']     - games_4['low_off_rtg']
matchups['net_rtg']     =   games_4['high_net_rtg']     - games_4['low_net_rtg']
matchups['pace']        =   games_4['high_pace']        - games_4['low_pace']
matchups['hhs_net_rtg'] =   games_4['high_hhs_net_rtg'] - games_4['low_hhs_net_rtg']
matchups['hhs_off_rtg'] =   games_4['high_hhs_off_rtg'] - games_4['low_hhs_def_rtg']
matchups['hhs_def_rtg'] =   games_4['high_hhs_def_rtg'] - games_4['low_hhs_off_rtg']
matchups['rpi']         =   games_4['high_rpi']         - games_4['low_rpi']
matchups['ppp']         =   games_4['high_ppp']         - games_4['low_pppd']
matchups['ppsa']        =   games_4['high_ppsa']        - games_4['low_opp_ppsa']
matchups['efg%']        =   games_4['high_efg%']        - games_4['low_efgd%']
matchups['ftr']         =   games_4['high_ftr']         - games_4['low_ftrd']
matchups['3pr']         =   games_4['high_3pr']         - games_4['low_3prd']
matchups['%pts_ft']     =   abs(games_4['high_%pts_ft'] - games_4['low_%pts_ftd'])
matchups['%pts_2p']     =   abs(games_4['high_%pts_2p'] - games_4['low_%pts_2pd'])
matchups['%pts_3p']     =   abs(games_4['high_%pts_3p'] - games_4['low_%pts_3pd'])
matchups['orb%']        =   games_4['high_orb%']        - (1-games_4['low_drb%'])
matchups['drb%']        =   games_4['high_drb%']        - (1-games_4['low_orb%'])
matchups['trb%']        =   games_4['high_trb%']        - games_4['low_trb%']
matchups['ast%']        =   games_4['high_ast%']        - games_4['low_astd%']
matchups['tov%']        =   games_4['high_tov%']        - games_4['low_tovd%']
matchups['ast_tov']     =   games_4['high_ast_tov']     - games_4['low_opp_ast_tov']
matchups['stl%']        =   games_4['high_stl%']        - games_4['low_opp_stl%']
matchups['blk%']        =   games_4['high_blk%']        - games_4['low_opp_blk%']
matchups['pf%']         =   games_4['high_pf%']         - games_4[ 'low_opp_pf%']
matchups['pppd']        =   games_4['high_pppd']        - games_4['low_ppp']
matchups['opp_ppsa']    =   games_4['high_opp_ppsa']    - games_4['low_ppsa']
matchups['efgd%']       =   games_4['high_efgd%']       - games_4['low_efg%']
matchups['ftrd']        =   games_4['high_ftrd']        - games_4['low_ftr']
matchups['3prd']        =   games_4['high_3prd']        - games_4['low_3pr']
matchups['%pts_ftd']    =   abs(games_4['high_%pts_ftd']- games_4['low_%pts_ft'])
matchups['%pts_2pd']    =   abs(games_4['high_%pts_2pd']- games_4['low_%pts_2p'])
matchups['%pts_3pd']    =   abs(games_4['high_%pts_3pd']- games_4['low_%pts_3p'])
matchups['astd%']       =   games_4['high_astd%']       - games_4['low_ast%']
matchups['tovd%']       =   games_4['high_tovd%']       - games_4['low_tov%']
matchups['opp_ast_tov'] =   games_4['high_opp_ast_tov'] - games_4['low_ast_tov']
matchups['opp_stl%']    =   games_4['high_opp_stl%']    - games_4['low_stl%']
matchups['opp_blk%']    =   games_4['high_opp_blk%']    - games_4['low_blk%']
matchups['opp_pf%']     =   games_4['high_opp_pf%']     - games_4['low_pf%']
```

```python
first_weekend = matchups[(matchups['round'] == 'First Round') | (matchups['round'] == 'Second Round')]
print(first_weekend.shape)

elite = matchups[(matchups['round'] == 'Sweet 16') | (matchups['round'] == 'Elite Eight') 
                | (matchups['round'] == 'Final Four') | (matchups['round'] == 'Championship')]
print(elite.shape)
```
