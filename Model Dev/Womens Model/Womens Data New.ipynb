```python
import pandas as pd

master = pd.read_csv('mastering_wncaat_teams_historical.csv')
master = master[master['year']>=2021]
games = pd.read_csv('mastering_wncaat_games_historical.csv')
games = games[games['year']>=2021]
torvik = pd.read_csv('raw_torvik_women_historical.csv')
torvik['adj_de'] = 200 - torvik['adj_de']
torvik = torvik[[
        'torvik_id',
        'adj_oe', 
        'adj_de', 
        'barthag', 
        'efg%', 
        'efgd%', 
        'tor', 
        'tord', 
        'orb%',
        'drb%', 
        'ftr', 
        'ftrd', 
        '2p%', 
        '2p%d', 
        '3p%', 
        '3p%d', 
        '3pr', 
        '3prd',
        'adj_tempo', 
        'wab']]
```

```python
print(master.shape)
print(torvik.shape)
df_teams = pd.merge(master, torvik, left_on='torvik_id', right_on='torvik_id')
print(df_teams.shape)
```

```python
print(games.shape)
df_games = pd.merge(games, df_teams, left_on='high_team_id', right_on='team_id')
print(df_games.shape)
df_games = df_games.rename(columns={
     'year_x': 'year',
     'region_x': 'region',
     'round': 'round',
     'adj_oe': 'high_adj_oe',
     'adj_de': 'high_adj_de',
     'barthag': 'high_barthag',
     'efg%': 'high_efg%',
     'efgd%': 'high_efgd%',
     'tor': 'high_tor',
     'tord': 'high_tord',
     'orb%': 'high_orb%',
     'drb%': 'high_drb%',
     'ftr': 'high_ftr',
     'ftrd': 'high_ftrd',
     '2p%': 'high_2p%',
     '2p%d': 'high_2p%d',
     '3p%': 'high_3p%',
     '3p%d': 'high_3p%d',
     '3pr': 'high_3pr',
     '3prd': 'high_3prd',
     'adj_tempo': 'high_adj_tempo',
     'wab': 'high_wab'
})
df_games = pd.merge(df_games, df_teams, left_on='low_team_id', right_on='team_id')
print(df_games.shape)
df_games = df_games.rename(columns={
     'year_x': 'year',
     'region_x': 'region',
     'adj_oe': 'low_adj_oe',
     'adj_de': 'low_adj_de',
     'barthag': 'low_barthag',
     'efg%': 'low_efg%',
     'efgd%': 'low_efgd%',
     'tor': 'low_tor',
     'tord': 'low_tord',
     'orb%': 'low_orb%',
     'drb%': 'low_drb%',
     'ftr': 'low_ftr',
     'ftrd': 'low_ftrd',
     '2p%': 'low_2p%',
     '2p%d': 'low_2p%d',
     '3p%': 'low_3p%',
     '3p%d': 'low_3p%d',
     '3pr': 'low_3pr',
     '3prd': 'low_3prd',
     'adj_tempo': 'low_adj_tempo',
     'wab': 'low_wab'
})
```

```python
df_games.columns.to_list()
```

```python
df_matchups = df_games[['year', 'region', 'round', 'high_bracket_seed', 'high_team_id', 'low_bracket_seed', 'low_team_id', 'win']]
df_matchups['adj_oe']      = df_games['high_adj_oe'] - df_games['low_adj_de']
df_matchups['adj_de']      = df_games['high_adj_de'] - df_games['low_adj_oe']
df_matchups['adj_barthag'] = df_games['high_barthag'] - df_games['low_barthag']
df_matchups['efg%']        = df_games['high_efg%'] - df_games['low_efgd%']
df_matchups['adj_efgd%']   = df_games['high_efgd%'] - df_games['low_efg%']
df_matchups['tor']         = df_games['high_tor'] - df_games['low_tord']
df_matchups['tord']        = df_games['high_tord'] - df_games['low_tor']
df_matchups['orb%']        = df_games['high_orb%'] - df_games['low_drb%']
df_matchups['drb%']        = df_games['high_drb%'] - df_games['low_orb%']
df_matchups['ftr']         = df_games['high_ftr'] - df_games['low_ftrd']
df_matchups['ftrd']        =  df_games['high_ftrd'] - df_games['low_ftr']
df_matchups['2p%']         =  df_games['high_2p%'] - df_games['low_2p%d']
df_matchups['2p%d']        = df_games['high_2p%d'] - df_games['low_2p%']
df_matchups['3p%']         = df_games['high_3p%'] - df_games['low_3p%d']
df_matchups['3p%d']        = df_games['high_3p%d'] - df_games['low_3p%']
df_matchups['3pr']         = df_games['high_3pr'] - df_games['low_3prd']
df_matchups['3prd']        = df_games['high_3prd'] - df_games['low_3pr']
df_matchups['adj_tempo']   = df_games['high_adj_tempo'] - df_games['low_adj_tempo']
df_matchups['wab']         = df_games['high_wab'] - df_games['low_wab']
```
